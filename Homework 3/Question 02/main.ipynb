### Self Organizing Map

```python
# initializing
from SOM import SOM
from dataloader import DataLoader
import numpy as np
```

We are using the data of CO2 global emmisions.

```python
CO2_data = DataLoader(data_path='co2-emissions-per-country.csv')
CO2_data.load_data()
```

```python
# Data is organized according to the year
CO2_data.data.keys()
```

```python
year_2020 = np.array(CO2_data.data['2020'])
num_rows = year_2020.shape[0]
num_cols = year_2020.shape[1]

print("Data Attributes: ", num_cols)
print("Countries: ", num_rows)
```

```python
year_2020[0:10]
```

```python
# for better results size of the map must be of the same size as labels
key = '2020'
map_size = (15, 15, num_cols-1)
som = SOM(map_size=map_size ,learning_rate=0.1,epochs=1000)
som.train(year_2020, with_labels=True, with_plot=False)
som.plot_map(save=True, filename="CO2_map"+key+".png")
som.save_map(filename="Maps/CO2_map"+key+".json")
```
